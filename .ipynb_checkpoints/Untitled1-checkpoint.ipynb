```python
model = timm.create_model('CBAMresnet50', pretrained=True)
model
```

```python
import torch
names=[]
kernel_sizes=[]
for name, module in model.named_modules():
    if isinstance(module, torch.nn.Conv2d):
        names.append(name)
        kernel_sizes.append(module.kernel_size[0])
        #print(name, module.kernel_size)
print(len(names))
```
